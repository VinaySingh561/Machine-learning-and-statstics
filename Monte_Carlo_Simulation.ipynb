```python
import numpy as np
import matplotlib.pyplot as plt
```

## Problem  - You are supposed to submit two reports to your manager , each will take 1 to 5 hours and 2 to 6 hours respectively. And you have to attend a crucial family function in 9 hours. Will you be able to make it on time?

```python
sims = 100000
report_1 = np.random.uniform(1,5,sims) ## first report
report_2 = np.random.uniform(2,6,sims) ## second report
duration = report_1 + report_2   ## total time on making reports
```

```python
plt.hist(duration)
plt.axvline(9,color = 'red')
plt.show()
print("Prob that you won't be able to make it on time {:.2f} %".format((duration > 9).sum()*100/sims))
```
